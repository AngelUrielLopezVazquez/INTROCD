Cambio e inputacion de variables en columnas con datos numericas

```python
columns_df = ['eventid', 'iyear', 'imonth', 'iday', 'extended', 
              'country', 'region', 'specificity', 'vicinity', 'crit1', 
              'crit2', 'crit3', 'doubtterr', 'multiple', 'success', 
              'suicide', 'attacktype1', 'targtype1', 'targsubtype1', 'natlty1', 
              'guncertain1', 'individual', 'weaptype1', 'property', 'ishostkid', 
              'INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY']

for col in columns_df:
    df1.dropna(subset=[col], inplace=True)
    df1[col] = df1[col].astype(float)
    df1[col] = df1[col].astype(int)

df1.head()
```

En las columnas iday e imonth, se remplaza el 0 por la media aritmetica de los datos de cada columna

```python
df1['iday'].replace(0, df1['iday'].mean(), inplace=True)
df1['imonth'].replace(0, df1['imonth'].mean(), inplace=True)
df1['iday'] = df1['iday'].astype(float)
df1['iday'] = df1['iday'].astype(int)
df1['imonth'] = df1['imonth'].astype(float)
df1['imonth'] = df1['imonth'].astype(int)
df1.head()
```

Con los datos de las columnas iyear, imonth e iday, se establece la fecha especifica del ataque y se coloca en la columna approxdate

```python
df1['approxdate'] = df1['iyear'].astype(str) + '/' + df1['imonth'].astype(str) + '/' + df1['iday'].astype(str)
df1['approxdate'] = pd.to_datetime(df1['approxdate'], errors='coerce')
df1.head()
```

En las columnas nperps, nkill y nwound, se remplazan los valores vacios con la media de los datos de cada columna

```python
columns_df = ['nperps', 'nkill', 'nwound']

for col in columns_df:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')
    df1[col].fillna(df1[col].mean(), inplace=True)
    df1[col] = df1[col].astype(float)
    df1[col] = df1[col].astype(int)

df1.head()
```

En las columnas 'weapsubtype1' y 'propextent' al ser variables si/no, se asume la inexistencia de esa informacion por lo que se coloca 0 en las columnas vacias

```python
lista = ['weapsubtype1', 'propextent']
for col in lista:
    df1[col].fillna(0, inplace=True)
df1.head()
```

Conversion de tipo de datos a numericos (enteros o flotantes en columnas especificas,) ademas de imputaciones especificas ('ransom')

```python
df1['weapsubtype1'] = df1['weapsubtype1'].astype(float)
df1['weapsubtype1'] = df1['weapsubtype1'].astype(int)

df1['propextent'] = df1['propextent'].astype(float)
df1['propextent'] = df1['propextent'].astype(int)

df1['ransom'].fillna(-1, inplace=True)
df1['ransom'] = df1['ransom'].astype(float)
df1['ransom'] = df1['ransom'].astype(int)

df1['latitude'] = df1['latitude'].astype(float)
df1['longitude'] = df1['longitude'].astype(float)

columns_to_fill = ['propextent_txt', 'weapsubtype1_txt']
df1.loc[:, df1.select_dtypes(include='int').columns] = df1.select_dtypes(include='int').replace(-9, -1)

for column in columns_to_fill:
    df1[column].fillna('Unknown', inplace=True)
df1.head()
```
